```python
import numpy as np
from collections import Counter
#counter is used to count the occurrences of elements in a list or iterable. in KNN after finding the k nearest neighbors, we can use Counter to determine the most common class among those labels counter helps count how many times each label appears.
```

```python
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))
```

```python
def knn_predict(training_data, training_labels, test_point, k):
    distances = []
    for i in range(len(training_data)):
        dist= euclidean_distance(test_point, training_data[i])
        distances.append((dist, training_labels[i]))
        distances.sort(key=lambda x:x[0])
        k_nearest_labels = [label for _, label in distances[:k]]
        return Counter(k_nearest_labels).most_common(1)[0]
#distances.append saves how far each training point is from the test point, along with its label.
#distances.sort is used to sorts the list so the nearest points come first.
#k_nearest_labels picks the labels of the k closest points.
#Uses Counter to find which label appears most among those k labels that becomes the prediction.
```

```python
training_data = [[1,2], [2,3], [3,1], [6,5], [7,7], [8,6]]
training_labels = [0, 0, 0, 1, 1, 1]
test_point = [4, 4]
k = 3
```

**Prediction**

```python
prediction = knn_predict(np.array(training_data), training_labels, np.array(test_point), k)
print(prediction)
```
